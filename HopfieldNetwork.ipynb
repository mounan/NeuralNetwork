```python
# 正方行列にする
def reshape(data):
    dim = int(np.sqrt(len(data)))
    data = np.reshape(data, (dim, dim))
    return data
```

```python
class HopfieldNetwork(object):
    #閾値
    threshold = 0
    
    
    def fit(self, train_data):
        self.dim = len(train_data[0])
        p = len(train_data)
        self.J = np.zeros((self.dim, self.dim))
        
        for m in range(p):
            t = train_data[m]
            tt = np.outer(t,t)
#             for i in range(20):
            self.J += tt
    
        for i in range(self.dim):
            self.J[i, i] = 0
            
        self.J /= self.dim
       
    #予測
    def predict(self, data, threshold=0, loop=10):
        """
        更新: 入力値のリストに対し、出力値のリストを返す
        """
        self.threshold = threshold
        return [self._predict(d, loop=loop) for d in data]

    def _predict(self, xr, loop=10):
        for i in range(loop):
            xr = self.J.dot(xr) - self.threshold
            # xr の符号をとる
            xr = np.sign(xr)
        return xr
         
        
    def plot_data(self, ax, data):
        """
        data を画像化 (2次元化) して ax にプロット
        """
        dim = int(np.sqrt(len(data)))
        # このサンプルで扱う画像は縦横 同じ長さのもののみ
        assert dim * dim == len(data)

        img = (data.reshape(dim, dim) + 1) / 2
        ax.imshow(img, cmap=cm.Greys_r, interpolation='nearest')
        return ax
    
def get_corrupted_input(input, corruption_level):
    corrupted = np.copy(input)
    inv = np.random.binomial(n=1, p=corruption_level, size=len(input))
    for i, v in enumerate(input):
        if inv[i]:
            corrupted[i] = -1 * v
    return corrupted


def plot(hn, data, test, predicted, figsize=(5, 7)):
    
    print("p = ",len(data))
    print("alpha = ", len(data)/len(data[0]))
    fig, axes = plt.subplots(len(data), 3, figsize=figsize)
    for i, axrow in enumerate(axes):
        if i == 0:
            axrow[0].set_title('train_data')
            axrow[1].set_title("test_data")
            axrow[2].set_title('output')
        hn.plot_data(axrow[0], data[i])
        hn.plot_data(axrow[1], test[i])
        hn.plot_data(axrow[2], predicted[i])

        for ax in axrow:
            ax.xaxis.set_visible(False)
            ax.yaxis.set_visible(False)
    return fig, axes
```
